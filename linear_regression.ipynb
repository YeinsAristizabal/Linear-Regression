```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gradient Descent

def funcion1_RLineal(X, y, w, b, omega, n):

    dw = -(1/n) * np.sum(X * (y - (w*X+b)))
    db = -(1/n) * np.sum(y - (w*X+b))

    w = w - omega * dw
    b = b - omega * db

    # plt.plot(X,y,'or',markersize=4)
    # plt.plot(X, (w*X+b))
    # plt.show()

    return w, b
```

```python
plt.subplots(figsize=(8,5))
X = np.linspace(0, 10, 50)
y = 3*X + 10 + (np.random.random(50) - 0.8) * 7
plt.scatter(X, y, s=30, c='lime')
plt.show()
```

```python
X = np.reshape(X, (len(X),1))
y = np.reshape(y, (len(y),1))
omega = 0.05
n_iter = 500
n = X.shape[0]

w = np.random.rand(X.shape[1], 1)
b = np.random.rand()

for i in range(n_iter):
  w, b = funcion1_RLineal(X, y, w, b, omega, n)

print(w, b)

plt.subplots(figsize=(8,5))
plt.plot(X,y,'or',markersize=4)
plt.plot(X, (w*X+b))
plt.show()
```
